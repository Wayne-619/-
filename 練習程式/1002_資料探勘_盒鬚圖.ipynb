```python
train_x['Sex']
le = pp.LabelEncoder()
le.fit(train_x['Sex'])
#print(le.classes_)
#print(le.transform(train_x['Sex']))
n_Sex = le.transform(train_x['Sex'])
n_Age = train_x['Age']
N_Sex = pd.DataFrame(n_Sex)
N_Age = pd.DataFrame(n_Age)
N_train = N_Sex.merge(N_Age, how='outer', left_index=True, right_index=True).dropna()
N_train.describe()
```
